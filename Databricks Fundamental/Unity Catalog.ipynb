# Row filtering by function

```sql
CREATE FUNCTION us_filter(region STRING)
RETURN IF(IS_MEMBER('admin'), true, region='US');

CREATE FUNCTION name_mask(name STRING)
RETURN CASE WHEN IS_MEMBER('admin') THEN NAME ELSE '****' END;


ALTER TABLE sales SET ROW FILTER us_filter ON region; -- apply filter to table
ALTER TABLE sales DROP ROW FILTER; -- remove row filter for table

ALTER TABLE users ALTER COLUMN name SET MASK name_mask; -- apply mask for the column
ALTER TABLE users ALTER COLUMN name DROP MASK; -- remove mask from column
DROP FUNCTION name_mask; -- remove function
```

```sql
CREATE VIEW view_name AS
SELECT
  CASE
    WHEN is_account_group_member('admin') THEN column_name
    ELSE 'REDACTED'
  END as column_name -- column mask
FROM
  'some_catalog'.'table_name'
WHERE
  CASE
    WHEN is_account_group_member('admin') THEN TRUE
    ELSE device_id < 30
    END; -- row filter
```
